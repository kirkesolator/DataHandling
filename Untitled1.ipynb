```python
# Test to pick out 10 values from each element that meets our criterium
signalV = np.random.rand(10000,1)
```

```python
fig = plt.figure(figsize=[12,7])
ax1 = fig.add_subplot(121) 
ax1.plot(signalV,'.k')
ax2 = fig.add_subplot(122) 
ax2.plot(signalV,'.k')
plt.show()
```

```python
plt.hist(signalV,200,color='k')
plt.show()
```

```python
# Hanne example to pull out window around trigger events:
wWidth = 50
tHold = 0.005
signalList,indices = zip(*[[signalV[index-wWidth:index+wWidth],index]
                          for index,value 
                          in enumerate(signalV) 
                          if value < tHold 
                          and len(signalV)-wWidth > index >= wWidth-1])

# We need matrix to plot
ourMat = np.zeros([len(signalList),wWidth*2])
for i in range(len(signalList)):
    ourMat[i,:] = [j for j in signalList[i]]
```

```python
plt.figure(figsize=[8,8])
plt.imshow(ourMat)
plt.show()
```

```python
plt.plot(indices,'.k')
plt.xlabel('Event')
plt.ylabel('')
plt.show()
```

```python
plt.plot(np.mean(ourMat,0),'k',linewidth=4)
plt.plot(np.mean(ourMat,0)-np.std(ourMat,0),color=[.7,.7,.7])
plt.plot(np.mean(ourMat,0)+np.std(ourMat,0),color=[.7,.7,.7])
plt.show()
```
